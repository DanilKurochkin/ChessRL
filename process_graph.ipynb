```python
import pandas as pd
```

```python
df_len = pd.read_csv('csv_data/mean_len.csv').drop(columns=['Wall time']).rename(columns={'Value': 'len'})
df_reward = pd.read_csv('csv_data/mean_reward.csv').drop(columns=['Wall time', 'Step']).rename(columns={'Value': 'reward'})
```

```python
df_res = pd.concat([df_len, df_reward], axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
#      (  )
fig, ax1 = plt.subplots(figsize=(10, 6))

#    (  Y)
ax1.plot(df_res['Step'], df_res['len'], color='tab:blue', label='Game length (MEAN)')
ax1.set_xlabel('Training Steps')
ax1.set_ylabel('Game Length', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.grid(True, linestyle='--', alpha=0.7)

#    ( )
ax2 = ax1.twinx()
ax2.plot(df_res['Step'], df_res['reward'], color='tab:red', label='Reward (MEAN)')
ax2.set_ylabel('Reward', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

#  
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='center left')

#   
plt.title('Game Length and Reward During Training')
plt.tight_layout()
plt.show()
```
